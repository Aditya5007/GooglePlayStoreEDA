```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("googleplaystore.csv")

data.head()
```

```python
data.info()
```

```python
data_df = pd.DataFrame(data)

```

```python
data_df.drop(['Android Ver'], axis=1 , inplace=True)

data_df
```

```python
data_df['Installs'] = data_df['Installs'].str.replace('+','')
data_df['Installs'] = data_df['Installs'].str.replace(',','')
data_df['Installs'] = pd.to_numeric(data_df['Installs'], errors='coerce')



```

```python
data_df.isna().sum()

na_rows = data_df[data_df.isna().any(axis=1)]

data_df.dropna(inplace=True)
```

```python
data_df.drop_duplicates(inplace=True)
```

```python
data_df
```

```python
category_count = data_df['Category'].value_counts()
total = category_count.sum()
threshold = 0.032
mask = (category_count / total) < threshold

# Replace categories with 'Others' and calculate their sum
category_count_masked = category_count[~mask]
category_count_masked['Others'] = category_count[mask].sum()

plt.pie(category_count_masked, labels=category_count_masked.index, autopct='%1.1f%%')
plt.title("Category Distribution")
plt.show()
```

```python
data_df
```

```python
average_rating = data_df.groupby('Category')['Rating'].mean()
top_5_average_rating = average_rating.sort_values(ascending=False).head(5)
plt.figure(figsize=(12,6))
plt.bar((top_5_average_rating.index),top_5_average_rating)

plt.xlabel('Category')
plt.ylabel('Average Rating')
plt.title('Average Rating of each Category')

plt.show()
```

```python
data_df['Reviews'] = pd.to_numeric(data_df['Reviews'], errors='coerce')
```

```python
filtered_df = data_df[data_df['Reviews'] > 50000]
average_rating = filtered_df.groupby('Category')['Rating'].mean()
total_reviews = filtered_df.groupby('Category')['Reviews'].sum()

modified_df = pd.DataFrame({'Average Rating': average_rating , 'Total Reviews' : total_reviews})
top_category = average_rating.sort_values(ascending=False).head(5)

modified_df.sort_values(by = 'Average Rating' , ascending =False).head(5)

```

```python
app_counts = data_df.groupby('Type')['App'].count()


plt.bar(app_counts.index, app_counts ,color = 'lightblue')
```

```python
data_df.drop(['Price'], axis =1 ,inplace =True)
```

```python
data_df['Size'].unique()

```

```python
data_df['Size'].replace('M' , 'e+6', regex=True, inplace=True)
data_df['Size'].replace('k', 'e+3' , regex=True, inplace=True)

data_df['Size'] = pd.to_numeric(data_df['Size'] , errors = 'coerce' )
```

```python
data_df.info()
```

```python
data_df['Size'].fillna(data_df['Size'].mean(), inplace=True)
```

```python
new_values = data_df['Size']/1000000

data_df['Size'] = new_values

data_df

```
